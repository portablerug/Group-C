```python
import numpy as np
import matplotlib.pyplot as plt

# Input the numbers as a list
input_numbers = [3.03053826,-6.94635893,8.97504555,10.6496473,10.43352994,7.97220345,-1.8502485,-15.30411813,-1.4261436,18.70463797]





# Calculate mean, median, variance, and standard deviation
mean = np.mean(input_numbers)
median = np.median(input_numbers)
variance = np.var(input_numbers)
std_deviation = np.std(input_numbers)

# Print the results
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_deviation)

# Create a histogram to visualize the data
plt.hist(input_numbers, bins=10, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Input Numbers')
plt.grid(True)

# Show the histogram
plt.show()
```
